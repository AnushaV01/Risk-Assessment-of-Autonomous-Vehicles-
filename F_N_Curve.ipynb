**F-N Curve**

Data Set: https://www.kaggle.com/datasets/ibriiee/tesla-autonomous-deaths-data-updated-2023#

According to Data: Number of Fatalities per Accident:
*   1 fatality: 248 accidents
*   2 fatalities: 40 accidents
*   3 fatalities: 7 accidents
*   4 fatalities: 3 accidents

Cumulative Frequency:
*   (Atleast 1 or more)1 fatality: 298 (248+40+7+3)
*   (Atleast 2 or more)2 fatalities: 50 (40+7+3)
*   (Atleast 3 or more)3 fatalities: 10 (7+3)
*   4 fatalities: 3


FN Curve Plot:
*   X-axis: Number of Fatalities per Accident - [1, 2, 3, 4]
*   Y-axis: Cumulative Frequency of Accidents - [298, 50, 10, 3]

```python
import matplotlib.pyplot as plt

# Provided data
number_of_fatalities = [1, 2, 3, 4]
frequency = [248, 40, 7, 3]

# Calculate the cumulative frequency in reverse order
cumulative_frequency = [sum(frequency[i:]) for i in range(len(frequency))]

# Plot the FN curve
plt.figure(figsize=(10, 6))
plt.plot(number_of_fatalities, cumulative_frequency, marker='o', linestyle='-')
plt.xlabel('Number of Fatalities per Accident')
plt.ylabel('Cumulative Frequency of Accidents')
plt.title('FN Curve for Tesla Autonomous Vehicle Fatalities')
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Provided data
number_of_fatalities = [1, 2, 3, 4]
frequency = [248, 40, 7, 3]

# Calculate the cumulative frequency in reverse order
cumulative_frequency = [sum(frequency[i:]) for i in range(len(frequency))]

# Convert to numpy arrays for plotting
x = np.array(number_of_fatalities)
y = np.array(cumulative_frequency)

# Plot the FN curve on a log-log scale
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number of Fatalities per Accident (log scale)')
plt.ylabel('Cumulative Frequency of Accidents (log scale)')
plt.title('FN Curve for Tesla Autonomous Vehicle Fatalities (Log-Log Scale)')
plt.grid(True, which="both", ls="--")
plt.show()
```
