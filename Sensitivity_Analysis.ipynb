```python
import matplotlib.pyplot as plt
import numpy as np

# Probability values
p = np.linspace(0, 1, 100)

# Equations for sensitivity analysis
equation1 = 37.5 * p + 75 * (1 - p)
equation2 = 78.75 * p + 157.5 * (1 - p)
equation3 = 105 * p + 30 * (1 - p)

# Create the plot
plt.figure(figsize=(10, 6))

plt.plot(p, equation1, label='$37.5p + 75(1-p)$')
plt.plot(p, equation2, label='$78.75p + 157.5(1-p)$')
plt.plot(p, equation3, label='$105p + 30(1-p)$')

# Adding labels and title
plt.xlabel('Probability (p)')
plt.ylabel('Expected Value')
plt.title('Sensitivity Analysis')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Probability values
p = np.linspace(0, 1, 1000)  # Increased number of points for better precision

# Equations for sensitivity analysis
equation1 = 37.5 * p + 75 * (1 - p)
equation2 = 78.75 * p + 157.5 * (1 - p)
equation3 = 105 * p + 30 * (1 - p)

# Intersection points
p1 = 0.4
p2 = 0.8293

# Values at intersection points
value1 = 37.5 * p1 + 75 * (1 - p1)
value2 = 105 * p2 + 30 * (1 - p2)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot sections of the equations
plt.plot(p[p <= p1], equation1[p <= p1], label='$37.5p + 75(1-p)$', color='blue')
plt.plot(p[(p > p1) & (p <= p2)], equation3[(p > p1) & (p <= p2)], label='$105p + 30(1-p)$', color='green')
plt.plot(p[p > p2], equation2[p > p2], label='$78.75p + 157.5(1-p)$', color='orange')

# Highlight the intersection points
plt.scatter([p1, p2], [value1, value2], color='red')
plt.text(p1, value1, f'({p1}, {value1:.2f})', color='red', ha='right')
plt.text(p2, value2, f'({p2}, {value2:.2f})', color='red', ha='right')

# Adding labels and title
plt.xlabel('Probability (p)')
plt.ylabel('Expected Value')
plt.title('Sensitivity Analysis')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
